```python
from os import listdir
from os.path import isfile, join
import json

products = {}

webs = [f for f in listdir('/home/sunji/EM_sigmod/2013_camera_specs')]
for w in webs:
    products[w] = {}
    items = [f for f in listdir('/home/sunji/EM_sigmod/2013_camera_specs/'+w)]
    for it in items:
        with open('/home/sunji/EM_sigmod/2013_camera_specs/'+w+'/'+it, 'r') as f:
            data = json.load(f)
            products[w][it.split('.')[0]] = data
        
```

```python
def combine(x):
    if type(x) == list:
        return ';'.join(x)
    else:
        return str(x)

import pandas as pd
ids = []
titles = []
models = []
brands = []
mpns = []
for source, items in products.items():
    for idx, item in items.items():
        titles.append(item['<page title>'].lower())
        ids.append(source + '//' +idx)
        local_models = []
        if 'model' in item:
            local_models.append(combine(item['model']).lower())
        if 'model id' in item:
            local_models.append(combine(item['model id']).lower())
        if 'family line' in item:
            local_models.append(combine(item['family line']).lower())
        if 'model no' in item:
            local_models.append(combine(item['model no']).lower())
        if 'series' in item:
            local_models.append(combine(item['series']).lower())
        models.append(';'.join(local_models))
        
        local_brands = []
        if 'manufacturer' in item:
            local_brands.append(combine(item['manufacturer']).lower())
        if 'brand' in item:
            local_brands.append(combine(item['brand']).lower())
        if 'manuf no' in item:
            local_brands.append(combine(item['manuf no']).lower())
        brands.append(';'.join(local_brands))
        
        if 'mpn' in item:
            mpns.append(combine(item['mpn']).lower())
        else:
            mpns.append('')
original_data = pd.DataFrame(data={'spec_id':ids, 'page_title':titles, 'camera_brand': brands, 'camera_model': models, 'camera_mpn': mpns})    
         
```
